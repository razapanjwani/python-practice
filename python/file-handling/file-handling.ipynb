```python
data = open("./abc.txt")
print(data.readline() ,end="") # "this is a generator function so it read the line one by one"
print(data.readline() ,end="")
print(data.readline() )
lines = data.readline()
print(type(lines),"tpyes")
# for line in data.readline():
#     display(line)
```

```python
data = open("./abc.txt")
lines = data.readlines()
print(lines[:1],"line")
for line in lines:
    print(line,end="")
```

```python
data = open("./abc.txt","r")
data.read()
# data.write("he is 17 years old") this will give error because the mode is set to read only
```

```python
data = open("./abc1.txt","w")
data.write("his age is 17.")
# data.read() this will give error because the mode is set to write only.
```

```python
# When we use open function the connectivity is made with the file so we should always close the connectivity. python 'with' block has made our life easy is diconnect automatically
data = open("./abc.txt","r+")
data.read()
data.write("his age is 17")
data.seek(3)
data.read()
data.close()
```

```python
with open("./abc.txt","a+") as file:
    file.write("he studies in college")
    file.read()
```

```python
with open("./abc.txt","a") as file:
    file.write("he studies in college named sm science")
```

```python
with open("./abc.txt","r") as file:
    print(file.read())
```
